```python
import numpy as np

x = np.random.rand(50)
x
```

```python
def formula(x):
    delta_T = 1/64

    sum_1 = 0
    for i in range(1, len(x)-1):
        value_1 = pow(
            ((x[i+1] - 2*x[i] + x[i-1])
             / pow(delta_T, 2)),
            2
        )

        sum_1 += value_1

    sum_2 = 0
    for i in range(1, len(x)-1):
        value_2 = pow(
            ((x[i-1] - x[i])
             / delta_T),
            2
        )

        sum_2 += value_2

    return sum_1 + sum_2/2
```

```python
import pandas as pd
import spacy

nlp = spacy.load("ru_core_news_sm")
```

```python
def text_cleaning(test_text):
    test_doc = nlp(test_text)
    test_doc = [token.lemma_ for token in test_doc if not token.is_punct]
    test_doc = [
        token for token in test_doc if not token in nlp.Defaults.stop_words]
    return ' '.join(test_doc)
```
