```python
table = df[df['shop_id']==0].groupby(['shop_id'], as_index=False).sum()
```

```python

sales_columns = table.columns[table.columns.str.contains(' Sales')]
revenue_columns = table.columns[table.columns.str.contains(' Revenue')]
dates = sorted(
    pd.to_datetime(
        sales_columns.str.split(' ').str[0],
        dayfirst=True)
)
```

```python

shop_sales = pd.DataFrame(
    index=dates,
    data=np.c_[
        table[sales_columns].T.values,
        table[revenue_columns].T.values
    ],
    columns=[
        'Sales',
        'Revenue'
    ]
)

shop_sales
```

```python
shop_sales.groupby(shop_sales.index.strftime("%a")).sum().iloc[[1, 5, 6, 4, 0, 2, 3], :]
```
